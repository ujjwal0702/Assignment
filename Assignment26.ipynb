# Answer 1:

Matplotlib:

Matplotlib is a popular and widely-used Python library for creating static, interactive, and animated visualizations. It provides a variety of plotting capabilities and is often used for data visualization tasks in scientific research, data analysis, and other fields. The primary module used for creating plots in Matplotlib is pyplot, which provides a simple interface for generating a wide range of plots.

Here are five types of plots that can be created using the pyplot module of Matplotlib:

Line Plot: A basic line plot is used to visualize the relationship between two variables. It shows data points connected by straight lines, making it easy to observe trends and patterns in the data.

Scatter Plot: A scatter plot displays individual data points as markers on a 2D plane, where each point represents the values of two variables. It is useful for identifying patterns or relationships between variables.

Bar Plot: A bar plot, also known as a bar chart, represents categorical data with rectangular bars of different lengths or heights. Each bar corresponds to a category, and the height of the bar indicates the value associated with that category.

Histogram: A histogram is used to visualize the distribution of a continuous variable. It divides the data into bins and displays the frequency or count of data points falling into each bin.

Pie Chart: A pie chart is a circular graph that shows the proportions of different categories relative to the whole. Each category is represented as a slice of the pie, with the area of the slice corresponding to the proportion of that category.

# Answer 2

A scatter plot is a type of data visualization that is used to display individual data points on a 2D plane, with each data point representing the values of two variables. The scatter plot is useful for visually identifying relationships or patterns between the two variables. It is particularly effective in showing how the two variables might be correlated.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
```

# Answer 3:

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display multiple plots in the same figure window, making it easy to compare different visualizations or data sets side by side.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

# Adjust layout to avoid overlapping titles
plt.tight_layout()

# Display the plots
plt.show()
```

# Answer 4:

A bar plot, also known as a bar chart, is a data visualization technique used to represent categorical data using rectangular bars. It is a simple way to compare different categories and display the relative values or frequencies associated with each category. The length or height of each bar is proportional to the value of the category it represents.

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

# Answer 5:

A box plot, also known as a box-and-whisker plot, is a statistical data visualization technique used to display the distribution of a dataset. It provides a graphical summary of the data's central tendency, dispersion, and potential outliers. The box plot is particularly useful for comparing multiple datasets or identifying differences in distributions.

```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)
```

```python
plt.boxplot(box2)
```
