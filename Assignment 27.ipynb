# Answer 1:

Scatter Plot:
Use: Scatter plots are used to visualize the relationship between two continuous variables. They help to identify patterns, trends, and correlations between the variables.

Line Plot:
Use: Line plots are suitable for displaying trends over time or continuous data points. They are commonly used for time series data and visualizing how a variable changes over an ordered sequence.

Bar Plot:
Use: Bar plots are used to display categorical data with rectangular bars, where the length of each bar represents the quantity of the data. They are useful for comparing different categories and their values.

Histogram:
Use: Histograms are used to visualize the distribution of a single continuous variable. They display the frequency or count of data points within predefined bins or intervals.

Heatmap:
Use: Heatmaps are used to represent the relationship between two categorical variables or two continuous variables through a color-encoded matrix. They are particularly useful for visualizing correlations in large datasets.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the flights dataset
flights_data = sns.load_dataset("flights")
# Reshape the data into a pivot table
flights_pivot = flights_data.pivot("month", "year", "passengers")
# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Flights Dataset Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
