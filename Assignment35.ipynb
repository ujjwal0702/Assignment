# Answer 1: Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

```python
import scipy.stats as stats

# Sample parameters
sample_mean = 50
sample_std_dev = 5
sample_size = 50  # Adjust this to the actual sample size

# Confidence level
confidence_level = 0.95

# Calculate the standard error of the mean (SEM)
sem = sample_std_dev / (sample_size ** 0.5)

# Calculate the margin of error (MOE) using the z-score for the given confidence level
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_score * sem

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

# Answer 2: Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (counts) of each color in the sample
observed_counts = np.array([18, 22, 19, 10, 11, 20])  # Replace these values with your actual counts

# Expected proportions (as decimals) for each color
expected_proportions = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20])

# Total number of M&Ms in the sample
total_sample_size = np.sum(observed_counts)

# Calculate the expected counts based on the expected proportions
expected_counts = expected_proportions * total_sample_size

# Perform the chi-square goodness of fit test
chi2, p_value = chisquare(observed_counts, f_exp=expected_counts)

# Define the significance level
alpha = 0.05

# Determine whether to reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis: The distribution of M&M colors does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The distribution of M&M colors matches the expected distribution.")
```

# Answer 4: A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

```python
import scipy.stats as stats

# Sample size (number of individuals in the population)
sample_size = 500

# Number of individuals who smoke in the sample
sample_smokers = 60

# Proportion of smokers in the sample
sample_proportion = sample_smokers / sample_size

# Confidence level
confidence_level = 0.95

# Calculate the standard error of the proportion
std_error = (sample_proportion * (1 - sample_proportion) / sample_size) ** 0.5

# Calculate the margin of error using the z-score for the given confidence level
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_score * std_error

# Calculate the confidence interval
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

print(f"95% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
```

# Answer 6: Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the degrees of freedom
df = 10

# Generate values for the chi-square distribution
x = np.linspace(0, 30, 1000)
y = stats.chi2.pdf(x, df)

# Create the plot
plt.plot(x, y, label=f'Chi-Square (df={df})', color='b')

# Shade the area corresponding to a chi-square statistic of 15
x_shade = np.linspace(0, 15, 1000)
y_shade = stats.chi2.pdf(x_shade, df)
plt.fill_between(x_shade, y_shade, color='gray', alpha=0.5, label='Chi-Square (df=10) > 15')

# Label the axes and add a legend
plt.xlabel('Chi-Square Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.title('Chi-Square Distribution with Shaded Area')
plt.grid()
plt.show()
```

# Answer 7:A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

```python
import scipy.stats as stats

# Sample parameters
sample_size = 1000
sample_preferred_coke = 520

# Proportion of people in the sample who preferred Coke
sample_proportion = sample_preferred_coke / sample_size

# Confidence level
confidence_level = 0.99

# Calculate the standard error of the proportion
sem = ((sample_proportion * (1 - sample_proportion)) / sample_size) ** 0.5

# Calculate the margin of error (MOE) using the z-score for the given confidence level
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_score * sem

# Calculate the confidence interval
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

print(f"99% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
```

# Answer 8: A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequency of tails
observed_tails = 45

# Expected frequency for tails (fair coin)
expected_tails = 50

# Calculate the chi-square statistic
chi2 = ((observed_tails - expected_tails) ** 2) / expected_tails

# Define the significance level
alpha = 0.05

# Calculate the p-value
p_value = 1 - stats.chi2.cdf(chi2, df=1)

# Determine whether to reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis: The coin may be biased towards tails.")
else:
    print("Fail to reject the null hypothesis: There is no evidence of bias towards tails.")
```

# Answer 11: A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

```python
import scipy.stats as stats

# Sample parameters
sample_mean = 72
sample_std_dev = 10
sample_size = 30
population_mean = 70

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / (sample_size ** 0.5))

# Define the significance level
alpha = 0.05

# Calculate the degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the critical t-value for a two-tailed test
critical_t_value = stats.t.ppf(1 - alpha / 2, degrees_of_freedom)

# Calculate the p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), degrees_of_freedom))

# Determine whether to reject the null hypothesis
if abs(t_statistic) > critical_t_value:
    print("Reject the null hypothesis: The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference from 70.")

print(f"t-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
```
