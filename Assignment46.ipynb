## Answer 5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.boxplot(x=df['Price'])
plt.title("Flight plice Boxplot")
plt.xlabel('Prices')
plt.grid(True)
plt.show()
```

In the boxplot, potential outliers are typically represented by data points that fall outside the whiskers of the plot, which are calculated based on the interquartile range (IQR). Points beyond 1.5 times the IQR above the upper quartile or below the lower quartile are often considered potential outliers.

Once you identify potential outliers, it's essential to consider how they may impact your analysis. Outliers can affect statistical measures such as the mean and standard deviation, leading to biased results. They may also distort visualizations and models, potentially skewing interpretations of the data.

Depending on the context, you may choose to handle outliers in various ways:

1. **Keep them:** If outliers represent genuine data points and are not due to errors, you might choose to keep them in the analysis.
2. **Remove them:** If outliers are due to errors or anomalies, you might choose to remove them from the dataset.
3. **Transform data:** You can transform the data (e.g., log transformation) to mitigate the impact of outliers.
4. **Use robust statistics:** Use statistical measures that are less sensitive to outliers, such as the median instead of the mean.

## Answer 12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'Varies with device' with NaN
df['Size'] = df['Size'].replace('Varies with device', pd.NA)

# Convert the 'Size' column to numeric, removing 'M' and 'k' suffixes
df['Size'] = df['Size'].str.replace('M', 'e6').str.replace('k', 'e3').str.replace('+', '').str.replace(',', '')

# Drop rows with missing values in 'Size' and 'Rating' columns
df = df.dropna(subset=['Size', 'Rating'])

# Convert 'Size' column to numeric
df['Size'] = pd.to_numeric(df['Size'])

# Create a scatter plot to visualize the relationship between app size and rating
plt.figure(figsize=(10, 6))
plt.scatter(df['Size'], df['Rating'], alpha=0.5)
plt.title('Relationship Between App Size and Rating')
plt.xlabel('App Size (bytes)')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

## Answer 13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
# clean the 'Price' column by removing the '$' sign and converting to numeric
df['Price']=df['Price'].apply(lambda x: float(x.strip('$')) if '$' in x else 0)

# Group the dataset by "Tips" and calculate the average price for each app type
avg_price_by_type=df.groupby('Type')['Price'].mean().reset_index()

# Create a bar chart to compare  average prices by app type
plt.figure(figsize=(10,6))
plt.bar(avg_price_by_type['Type'],avg_price_by_type['Price'],color='skyblue')
plt.title('Average Prices by App type')
plt.xlabel("App type")
plt.ylabel("Average Price ($)")
plt.xticks(rotation=45, ha="right")
plt.grid(axis='y')
plt.show()
```

## Answer 14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
import pandas as pd

# Remove any duplicate entries based on 'App' column, keeping the entry with the highest number of reviews
df = df.sort_values('Reviews', ascending=False).drop_duplicates('App')

# Remove non-numeric characters from the 'Installs' column
df['Installs'] = df['Installs'].str.replace('[^\d]', '', regex=True)

# Replace empty strings with NaN and then convert to integer
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

# Drop rows with NaN values in the 'Installs' column
df = df.dropna(subset=['Installs'])

# Sort the dataset by 'Installs' column in descending order
df =df.sort_values('Installs', ascending=False)

# Get the top 10 most popular apps based on the number of installs
top_10_apps = df[['App', 'Installs']].head(10)

print("Top 10 Most Popular Apps:")
print(top_10_apps)
```
